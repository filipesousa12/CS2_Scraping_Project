```python
#Get the maps+stats table per player
matchstats = browser.find_element(By.CLASS_NAME, 'matchstats')

#Gets the content of the stats table
stats = matchstats.find_elements(By.CLASS_NAME, 'stats-content')


```

```python
#Gets the maps to use as matching for the stats
maps_scrap = matchstats.find_element(By.CSS_SELECTOR, '.box-headline.flexbox.nowrap.header').find_elements(By.CLASS_NAME,'small-padding')

# Create a list to hold the extracted data
data = []

for i in range(len(maps_scrap)):
    # Find all 'stats-menu-link' elements within maps_scrap[i]
    menu_links = maps_scrap[i].find_elements(By.CLASS_NAME, 'stats-menu-link')
    
    for link in menu_links:
        map_name =link.find_elements(By.XPATH, "./*")[1].text
        map_id =link.find_elements(By.XPATH, "./*")[1].get_attribute('id')
        # Append the data as a dictionary to the list
        data.append({'map_name': map_name, 'id': map_id})
# Convert the list to a DataFrame
maps_table = pd.DataFrame(data)
```
